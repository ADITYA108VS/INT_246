```python
def rname(x):
  m=np.argmax(x)
  n=np.max(x)
  print(n)
  name=""
  if m==0:
    name= "bottle"
  elif m==1:
    name= "cat"
  elif m==2:
    name= "dog"
  elif m==3:
    name= "human"
  elif m==4:
    name="laptop"
  elif m==5:
    name= "smartphone"
  else:
    name= ""
  return (n,name)
```

```python
def predict2(img,model) :
    im=np.asarray(img)
    im=np.asarray([im])
    c,bnd = model.predict(im)
    bnd = np.round(bnd)
    bnd = bnd.astype('int64')
    probability,name=rname(c)
    if probability>0.99 and probability<=1:
         img = cv2.rectangle(img,(bnd[0][0],bnd[0][1]),(bnd[0][2],bnd[0][3]),(0,255,0),2)
         (w, h), _ = cv2.getTextSize(name, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)
         img = cv2.putText(img, name, (bnd[0][0], bnd[0][1]-5),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 1)
    return img
    
```

```python
video=cv2.VideoCapture(0)
while True:
    ret,frame=video.read()
    img=cv2.resize(frame,(220,220))
    k=predict2(img,model)
    k=cv2.resize(k,(1280,720))
    cv2.imshow('Frames',k)
    key=cv2.waitKey(10)
    if key == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
