FUNCTION FOR CREATING DATASET

```python

def crd(path):
  images=[]
  labels=[]
  dim=[]
  names=[]
  c=0
  for i in sorted(os.listdir(path)):
    for j in sorted(os.listdir(path+r'/'+i)):
      if j.endswith('.xml'):
        obj=minidom.parse(path+r'/'+i+r'/'+j)
        xmin = int(obj.getElementsByTagName('xmin')[0].firstChild.nodeValue)
        ymin =int(obj.getElementsByTagName('ymin')[0].firstChild.nodeValue)
        xmax=int(obj.getElementsByTagName('xmax')[0].firstChild.nodeValue)
        ymax=int(obj.getElementsByTagName('ymax')[0].firstChild.nodeValue)
        name=obj.getElementsByTagName('name')[0].firstChild.nodeValue
        dim.append([xmin,ymin,xmax,ymax])
        names.append(name)
        labels.append(int(c))
      else:
        img=cv2.imread(path+r'/'+i+r'/'+j)
        images.append(img)
    c=c+1

  return (np.asarray(images),np.asarray(labels),np.asarray(dim),np.asarray(names))
```

CREATING DATASET

```python
images,labels,dim,names=crd('/content/dataset')
```

***CONVERTING TO CATEGORICAL DATA***

```python
labels=tf.keras.utils.to_categorical(labels)
```

**FUNCTIONAL CNN MODEL FOR CLASIFICATION AS WELL AS BOUDING BOX REGRESSION**

```python
inp = tf.keras.layers.Input(shape=(220,220,3))
res = tf.keras.applications.ResNet152V2(include_top=False,weights="imagenet") (inp)
flatten = tf.keras.layers.Flatten() (res)
d1 = tf.keras.layers.Dense(128,activation='relu') (flatten)
d2 = tf.keras.layers.Dense(128,activation='relu') (d1)
out1 = tf.keras.layers.Dense(6,activation='softmax') (d2)
d3 = tf.keras.layers.Dense(256,activation='relu') (d2)
d4 = tf.keras.layers.Dense(256,activation='relu') (d3)
out2 = tf.keras.layers.Dense(4) (d4)
```

***CREATING THE MODEL***

```python
model=keras.models.Model(inputs=inp,outputs=[out1,out2])
```

***COMPILING THE MODEL***

```python
model.compile(optimizer='adam',loss=['categorical_crossentropy','huber'],metrics=['accuracy'])
```

***UTILITY FUNCTION***

```python
def rname(x):
  m=0
  m=np.argmax(x)
  print( m)
  if m==0:
    return "bottle"
  elif m==1:
    return "cat"
  elif m==2:
    return "dog"
  elif m==3:
    return "human"
  elif m==4:
    return "laptop"
  elif m==5:
    return "smartphone"
  else:
    return "not found"
```

***FUNCTION FOR PASSING IMAGE URL OR FILE AND CHECKING PREDICTION***

```python
def predict(img_path,model,url=False) :
    if url :
        pred_img = io.imread(img_path)
    else :
        pred_img = cv2.imread(img_path)
    pred_img = cv2.resize(pred_img,(220,220))
    c,bnd = model.predict(np.asarray([pred_img]))
    bnd = np.round(bnd)
    bnd = bnd.astype('int64')
    img = cv2.rectangle(pred_img,(bnd[0][0],bnd[0][1]),(bnd[0][2],bnd[0][3]),(0,255,0),2)
    print(c)
    name = rname(c)
    # For the text background
    print("it is a " + name)
    # Finds space required by the text so that we can put a background with that amount of width.
    (w, h), _ = cv2.getTextSize(name, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)
    # For printing text
    img = cv2.putText(img, name, (bnd[0][0], bnd[0][1]-5),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 1)
    plt.imshow(img)
    plt.show()
```

***FITTING THE MODEL THE IMAGES AND LABELS***

***Mine cnn model***

```python

inp2 = tf.keras.layers.Input(shape=(220,220,3))
cnn=keras.models.Sequential([
                             keras.layers.Conv2D(100,kernel_size=(3,3),input_shape=(220,220,3)),
                             keras.layers.Conv2D(120,kernel_size=(3,3),activation='relu'),
                             keras.layers.MaxPool2D(2,strides=2),
                             keras.layers.Conv2D(100,activation='relu',kernel_size=(3,3)),
                             keras.layers.Conv2D(140,activation='relu',kernel_size=(3,3)),
                             keras.layers.MaxPool2D(2,strides=2)

])(inp2)
flatten2 = tf.keras.layers.Flatten() (cnn)
d01 = tf.keras.layers.Dense(128,activation='relu') (flatten2)
d02 = tf.keras.layers.Dense(128,activation='relu') (d01)
out01 = tf.keras.layers.Dense(6,activation='softmax') (d02)
d03 = tf.keras.layers.Dense(256,activation='relu') (d02)
d04 = tf.keras.layers.Dense(256,activation='relu') (d03)
out02 = tf.keras.layers.Dense(4) (d04)
```

***CREATING MY CNN MODEL***

```python
model2=keras.models.Model(inputs=inp2,outputs=[out01,out02])
```

***COMPILING CNN MODEL***

```python
model2.compile(optimizer='adam',loss=['categorical_crossentropy','huber'],metrics=['accuracy'])
```

***FITTING THE MODEL***

```python
model2.fit(images,[labels,dim],epochs=100)
```
